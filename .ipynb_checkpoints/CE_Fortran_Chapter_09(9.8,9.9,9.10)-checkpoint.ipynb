```python
# include packages needed
import numpy as np
from discretize_tools import rouwenhorst, grow_grid
from scipy.interpolate import interp1d, interp2d
from scipy.optimize import root, minimize
import matplotlib.pyplot as plt
from simulate import simulate_AR
plt.rcParams["figure.figsize"] = (11, 5)
from linear import linint
import math as m
import time
```

```python
# Program 9.8 The heterogeneous agent model

class hetero:
    
    def __init__(self,
                γ = .5,       # intertemporal elasticity of substitution
                β = .96,      # discount factor
                α = .36,      # capital share
                δ = .08,      # depreciation
                ρ = .6,       # persistence
                n = 1000,     # number of assets (a) gridpoint
                m = 7,        # number of shocks
                lower = 0,    # a lower bound
                upper = 100,  # a upper bound
                grow = .01,   # growth rate of a
                T = 100000):  # number of periods
        self.γ, self.β, self.α, self.δ, self.ρ, self.n, self.m = γ, β, α, δ, ρ, n, m # setting parameters
        self.σe = 0.04*(1-ρ**2)
        self.al, self.au = lower, upper # asset bounds
        self.grow = grow
        self.a = grow_grid(lower, upper, grow, n = n) # set asset grids
        self.η, self.π, self.weight= rouwenhorst(ρ,self.σe,m, weight = True) # set shock parameters
        self.η = np.exp(self.η)
        self.L = np.sum(self.η * self.weight) # labour demand as labour supply
        self.ηt, self.ind = simulate_AR(self.π, self.η, T = T, index = True) # simulate wage shocks for transition path
        self.r, self.w, self.K, self.Y, self.c = self.price(verbose = False) # calculate interest rate, wage rate, capital, output, consumption
        self.at, self.ct, self.lead = self.simulate_path(r_input = self.r, w_input = self.w, c_input = self.c) # transition path of assets, consumption, and 

        
    def price(self, tol = 1e-5, itermax = 10000, verbose = False):
        """This property calculates the interest rate, wage rate, capital level, output level, and consumption"""
        global c_old, c_new, ite2
        # setting parameters
        η = self.η
        a = self.a
        L = self.L
        
        α = self.α
        β = self.β
        δ = self.δ
        ite2 = 0
        #initialize c_old and c_new
        c_old = np.zeros((self.n,self.m))
        for j in range(self.m):
            c_old[:,j] = 0.04 * a + η[j]
        c_new = c_old.copy()
        
        def asset_price(r):
            """This function calculates the market clearing conditions"""
            global ite2
            ite2 += 1
            K = (α / (r+δ) )**(1 / (1-α) )*L # aggregate capital
            w = (1-α)*( K/L )** α # wage rate
            Y = K**α*L**(1 - α) # output
            
            c = self.policy(r_input = r, w_input = w, verbose = False) # consumption
            at, ct, lead = self.simulate_path(r_input = r, w_input = w, c_input = c) # simulate assets and consumption transition

            A = np.sum(at[lead:])/len(at[lead:]) # aggregate assets
            clear = (A - K)/A # market clearing condition (when clear = 0)
            
            if verbose:
                print('Out: Iteration = {}, error = {:.6e}'.format(ite2, abs(clear[0])))
            return clear[0]
        # calculates market clearing conditions
        r = root(asset_price, x0 = 0.04162896, tol = tol).x[0] # interest rate
        K = (self.α/(r+self.δ))**(1/(1-self.α))*L # capital
        w = (1-self.α)*(K/L)**self.α # wage rate
        Y = K**self.α*L**(1-self.α) # output
        
        return r, w, K, Y, c_old
        
    def policy(self, r_input, w_input, tol = 1e-7, itermax = 10000, verbose = False):
        """This property calculates the optimal consumption level"""
        global c_old
        # setting parameters
        a = self.a # asset
        η = self.η # wage shocks
        r = r_input[0] # exogenous interest rate
        w = w_input # exogenous wage rate
        if verbose:
            print('r = {}'.format(r))
        
        def foc(c):
            """This function calculates first-order condition for this model"""
            aplus = (1 + r)*a[i] + w * η[j] - c # law of motion for assets
            il, ir, ϕ = linint(aplus, self.al, self.au, self.grow, self.n) # calculate aplus linear interpolation
            #il: left interpolation point ir: right interpolation point
            foc = c - ϕ*RHS[il,j] - (1-ϕ)*RHS[ir,j]
            return foc
        # initialize RHS
        RHS = np.zeros((self.n, self.m))
        for ite in range(1,itermax):
            for i in range(self.n):
                for j in range(self.m):
                    RHS[i,j] = (self.β*(1+r)*(self.π[j,:]@(c_old[i,:]**(-1/self.γ))))**(-self.γ) # equation 9.20 on p.380
            
            for j in range(self.m):
                for i in range(self.n):
                    c_new[i,j] = root(foc,x0 = c_old[i,j]).x[0] # optimal consumption
                    if c_new[i,j] > ((1+r)*a[i] + w*η[j]-self.al): # check for borrowing constraint
                        c_new[i,j] = ((1+r)*a[i] + w*η[j]-self.al)
            con_lev = np.amax(np.abs(c_new-c_old)/np.amax(np.abs(c_old))) # get convergence level
            if verbose:
                print('In: Iteration = {}, error = {:.6e}'.format(ite, con_lev))
            if con_lev < tol:
                break
            else:
                c_old = c_new.copy()
        
        return c_old

        
    def simulate_path(self, r_input, w_input, c_input, T = 100000, lead = 5000):
        """This property calculates the transtion path of consumption and assets"""
        # initialize variables
        c = c_input
        r, w = r_input, w_input
        
        ηt = self.ηt
        at = np.zeros(T)
        ct = np.zeros(T)
        
        #initial period
        at[0] = 1
        il, ir, ϕ = linint(at[0], self.al, self.au, self.grow, self.n) # calculate aplus linear interpolation
        ct[0] = ϕ*c[il,self.ind[0]] + (1-ϕ)*c[ir,self.ind[0]]  # calculate future consumption
        
        #transition period
        for t in range(1,len(at)):
            at[t] = min(w*ηt[t-1] + (1+r)*at[t-1] - ct[t-1], self.au) #
            il, ir, ϕ = linint(at[t], self.al, self.au, self.grow, self.n)
            ct[t] = ϕ*c[il,self.ind[t]] + (1-ϕ)*c[ir,self.ind[t]]
            
        return at, ct, lead
```

```python
# Program 9.9 Using the distribution function to create output

class hetero_dis:
    
    def __init__(self,
                γ = .5,      #intertemporal elasticity of substitution
                β = .96,     #discount factor
                α = .36,     #capital share
                δ = .08,     #depreciation rate
                ρ = .6,      #persistence
                n = 1000,    #number of asset grids
                m = 7,       #number of wage shocks
                lower = 0,   #lower bound of a
                upper = 100, #upper bound of a
                grow = .01,  #asset growth rate
                T = 100000,  #total period
                ση = 0.04):  #st.d. of wage shocks
        self.γ, self.β, self.α, self.δ, self.ρ, self.n, self.m = γ, β, α, δ, ρ, n, m
        self.σe = ση*(1-ρ**2) #st.d. of noise e
        self.al, self.au = lower, upper #asset lower and upper bound
        self.grow = grow #asset growth rate
        self.a = grow_grid(lower, upper, grow, n = n) #generate asset sequence
        self.η, self.π, self.weight= rouwenhorst(ρ,self.σe,m, weight = True) #middle step of generating wage shocks
        self.η = np.exp(self.η) #wage shocks
        self.L = np.sum(self.η * self.weight) #labour demand and labour supply
        self.ηt, self.ind = simulate_AR(self.π, self.η, T = T, index = True) #simulate wage shocks for transition path
        self.r, self.w, self.K, self.Y, self.c = self.price(verbose = False)
        self.ϕ = self.distribution(r_input = self.r, w_input = self.w, c_input = self.c) #distribution function

        
    def price(self, tol = 1e-5, itermax = 10000, verbose = False):
        """This property calculates the interest rate, wage rate, capital level, output level, and consumption"""
        global c_old, c_new, ite2, a, η, al, ar, φ, ϕ1, ϕ_new
        #setting parameters
        η = self.η
        a = self.a
        L = self.L
        
        α = self.α
        β = self.β
        δ = self.δ
        
        #initialize variables
        ite2 = 0
    
        c_old = np.zeros((self.n,self.m))
        for j in range(self.m):
            c_old[:,j] = 0.04 * a + η[j]
        c_new = c_old.copy()
        
        al = np.zeros((self.n,self.m))
        ar = np.zeros((self.n,self.m))
        φ = np.zeros((self.n, self.m))
        
        ϕ1 = np.ones((self.n,self.m)) * (1/(self.n*self.m))

        
        def asset_price(r):
            """This function calculates the market clearing conditions"""
            global ite2
            ite2 += 1
            K = (α / (r+δ) )**(1 / (1-α) )*L  #aggregate capital
            w = (1-α)*( K/L )** α # wage rate
            Y = K**α*L**(1 - α) # output
            r = r[0] # interest rate
            
            c = self.policy(r_input = r, w_input = w, verbose = False) # consumption
            ϕ1 = self.distribution(r_input = r, w_input = w, c_input = c, verbose = False) #distribution function

            A = np.sum(a*np.sum(ϕ1, axis = 1)) #aggregate asset
            clear = (A - K)/A #market clearing condition
            
            if verbose:
                print('Out: Iteration = {}, error = {:.6e}'.format(ite2, abs(clear[0])))
            return clear[0]
        
        r = root(asset_price, x0 = 0.04162896, tol = tol).x[0] # find the interest rate that clears the market
        K = (self.α/(r+self.δ))**(1/(1-self.α))*L # aggregate capital
        w = (1-self.α)*(K/L)**self.α # wage rate
        Y = K**self.α*L**(1-self.α) # output
        
        return r, w, K, Y, c_old
        
    def policy(self, r_input, w_input, tol = 1e-7, itermax = 10000, verbose = False):
        """This property calculates the optimal consumption level"""
        global c_old
        #initialize variables
        a = self.a  #asset
        η = self.η  #wage shocks
        r = r_input #interest rate
        w = w_input #wage rate
                
        def foc(c):
            """This function calculates first-order condition for this model"""
            aplus = (1 + r)*a[i] + w * η[j] - c #law of motion for asset savings
            
            il, ir, ϕ = linint(aplus, self.al, self.au, self.grow, self.n) # determine the gridpoints in between this decision lies
    
            foc = c - ϕ*RHS[il,j] - (1-ϕ)*RHS[ir,j]        #first-order condition   

            return foc
        # calculate RHS
        RHS = np.zeros((self.n, self.m))
        for ite in range(1,itermax):
            for i in range(self.n):
                for j in range(self.m):
                    RHS[i,j] = (self.β*(1+r)*(self.π[j,:]@(c_old[i,:]**(-1/self.γ))))**(-self.γ) # equation 9.20 on p.380
            
            for i in range(self.n):
                for j in range(self.m):
                    c_new[i,j] = root(foc,x0 = c_old[i,j]).x  # optimal consumption
                    if c_new[i,j] > ((1+r)*a[i] + w*η[j]-self.al): # check for borrowing constraint
                        c_new[i,j] = ((1+r)*a[i] + w*η[j]-self.al)
            con_lev = np.amax(np.abs(c_new-c_old)/np.amax(np.abs(c_old))) # get convergence level
            if verbose:
                print('In: Iteration = {}, error = {:.6e}'.format(ite, con_lev))
            if con_lev < tol or ite == (itermax -1):
                break
            else:
                c_old = c_new.copy()
        
        return c_old

        
    def distribution(self, r_input, w_input, c_input, tol = 1e-7, itermax = 10000, verbose = False):
        """This function calculates the distribution function """
        global φ, ϕ1, ϕ_new
        #initialize variables
        c = c_input #consumption
        r = r_input #interest rate
        w = w_input #wage rate
        
        if verbose:
            print('r = {}'.format(r))
        
        for i in range(self.n):
            for j in range(self.m):
                # calculate where this guy would go
                aplus = (1+r) *a[i] + w*η[j] - c[i,j] #aplus equation on p.388
                aplus = min(max(aplus, self.al), self.au)
                # determine the gridpoints in between this decision lies
                al[i,j], ar[i,j], φ[i,j] = linint(aplus, self.al, self.au, self.grow, self.n)
                #φ is the fraction al takes
        
        # iterate until the distribution function converges
        for ite in range(1, itermax):
            ϕ_new = np.zeros((self.n, self.m))
            for i in range(self.n):
                for j in range(self.m):
                        ϕ_new[int(al[i,j]),:] += self.π[j,:] * (φ[i,j]*ϕ1[i,j])  #equation ϕ on p.388
                        ϕ_new[int(ar[i,j]),:] += self.π[j,:] * ((1-φ[i,j])*ϕ1[i,j]) #equation ϕ on p.388
                
            con_lev = np.amax(np.abs(ϕ_new-ϕ1)/np.amax(np.abs(ϕ1))) # get convergence level
            if verbose:
                print('Iteration = {}, error = {:.5e}'.format(ite, con_lev))
            # check for convergence
            if con_lev < tol or ite ==( itermax - 1):
                break
            else: # update distribution
                ϕ1 = ϕ_new.copy()
                
        return ϕ1
```

```python
#count for time use
tstart = time.time() # time that starts

model4 = hetero_dis() # run for the model

tend = time.time() # time that ends

time_taken = tend - tstart # time spent

print('Time used: {:.2f} minutes'.format(time_taken/60))
```

```python
# setting parameters and variables
ϕ = model4.ϕ  # distribution function
C = (model4.c * ϕ).sum() # aggregate consumption
Y = model4.Y # output
K = model4.K # aggregate capital
I = model4.δ*K # aggregate investment
r = model4.r # interest rate


η = model4.η # wage shocks
c = model4.c # consumption
a = model4.a # savings
L = model4.L # labour supply
C = (ϕ*c).sum() # aggregate consumption

# initialize variables
CVη = 0
CVc = 0
CVa = 0

A = np.sum(a*np.sum(ϕ, axis = 1))

# calculate coefficient of variations for shocks, consumption and savings
for i in range(model4.n):
    for j in range(model4.m):
        #nominator part of CV
        CVη += ϕ[i,j]*(η[j] - L)**2
        CVc += ϕ[i,j]*(c[i,j] - C)**2
        CVa += ϕ[i,j]*(a[i] - A)**2

CVη = CVη**0.5/L*100  #CV of wage shocks
CVc = CVc**0.5/C*100  #CV of consumption
CVa = CVa**0.5/A*100  #CV of savings
```
