# 3. The static general equilibrium model

In this chapter, we start the discussion with the most simple model structure and then successively introduce additional complexities, like government activity, intermediate goods production, or international trade.

# 3.1 The basic economy model
# 3.1.1 The command optimum.
Household: Suppliying capital $\overline{K}$, labour $\overline{L}$. Consuming good 1 and 2 which denoted by $X_1$ and $X_2$.

Firms: Using $\overline{K}$ and $\overline{L}$ to produce output $Y_1$ and $Y_2$ of good 1 and 2. $Y_i=L_i^{\beta_i}K_i^{1-\beta_i}, \quad i\in{1,2}$

Household utility function: $U(X_1,X_2)=X_1^\alpha X_2^{1-\alpha}$.

The economic problem is to allocate $K$ and $L$ to the producing of these two goods in order to produce an output combination which maximizes the utility of the representative household. The latter is called an <i>efficient allocation<i>.
    
The maximizing problem is given by:
\begin{align*}
    \max\limits_{X_1,X_2}X_1^{\alpha}X_2^{1-\alpha}\quad\text{s.t.}\quad & X_1=L_1^{\beta_1}K_1^{1-\beta_1}=Y_1,\quad X_2=L_2^{\beta_2}K_2^{1-\beta_2}=Y_2 \\
    & \overline{L}=L_1+L_2 \quad \text{and} \quad \overline{K}=K_1+K_2    
\end{align*}
    
For the formal derivation of the command optimum the constraints are substituted into the objective function so that the optimization problem becomes:
\begin{equation}
    \max\limits_{L_1,K_1}U(L_1,K_1)=\left[L_1^{\beta_1}K_1^{1-\beta_1}\right]^{\alpha}\left[(\overline{L}-L_1)^{\beta_2}(\overline{K}-K_1)^{1-\beta_2}\right]^{1-\alpha}
\end{equation}


```python
##### Program 3.1 Planner solution to the static equilibrium model

# Import package
# Using build-in function 'scipy.optimize.minimize' to find the maximum of the objective function
# Note this build-in function finds the minimization of scalar function of one or more variables.
import numpy as np
from scipy.optimize import minimize 

# Parameters
Kbar = 10 # Total capital is 10
Lbar = 20 # Total labour is 20
α = 0.3
β = ([0.3,0.6]) # β1 is the first element of β, β2 is the sencond element of β
K, L = np.zeros(2), np.zeros(2)

# Initial guess

# Define the objective utility function
# Note here we find the minimization of -U(L1,K1), this is equivalent to find the maximization of U(L1,K1)

def utility(x):
    """
    This defines above Cobb-Douglas utility function.
    The input x is an array of which the first element is K1
    and the second element is L1
    """
    utility = -((x[1]**β[0]*x[0]**(1-β[0]))**α*((Lbar-x[1])**β[1]*(Kbar-x[0])**(1-β[1]))**(1-α))
    return utility


# Initial Guess
x0 = np.ones(2) * 5 
# x0 is an array with first element is the initial guess
# for L1 and the second element is the initial guess for K1.

# Minimization routine
res = minimize(utility, x0)
# res is the optimize result
print(res)
# Solution
K[0] = res.x[0] # The first element of the solution array is K1
L[0] = res.x[1] # The second element of the solution array is L1
K[1] = Kbar - K[0] # Using resource constraint to calculate K2
L[1] = Lbar - L[0] # Using resource constraint to calculate L2
Y    = (L ** β) * (K ** (np.ones(2)-β)) # Using element-wise oprations to calculate the x1 (Y1) and x2 (Y2).

# Output
print('Goods Market 1 :')
print('X1 = ', Y[0], 'Y1 = ', Y[0])
print('Goods Market 2 :')
print('X2 = ', Y[1], 'Y2 = ', Y[1])
print('Labour Market :')
print('L1 = ', L[0], 'L2 = ', L[1], 'L = ', Lbar)
print('Capital Market :')
print('K1 = ', K[0], 'K2 = ', K[1], 'K = ', Kbar)
```

# 3.1.2 The market solution

In the previous subsection, we assumed that some central planner knows the techonologies, prefrences, and resources endowments. Of course, in reality there is no such central pllaner. Usually households know their own endowments and preferences and firms know their applied techonologies, however not all agents know everything about the endowments, prefrences, and techonologies fo their respective competitors. However, the interaction of supply and demand on goods and factor markets leads to an efficient allocation of resources despite the fact that the decisions of economic agents are decentralized. The drving force behind this result is the *price system*.
    
   In the present model, we assume $p_i$ as conumer prices and $q_i$ as producer prices for the two consumer goods. Assuming $w$ as the wage of labour supply and $r$ as the return of capital inputs. Furethermore, we assume perfect competition, i.e. prices are the same for all consumers and firms in the model and firms do not make profits.
     
    
   Given consumer and factor prices, the representative household maximizes utility by choosing consumption demand $X_i$ and keeping in mind the budget constraint:
\begin{equation}
    \max\limits_{X_1,X_2} X_1^\alpha X_2^{1-\alpha}\quad s.t. \quad p_1X_1+p_2X_2=w\overline{L}+r\overline{K}：=\overline{Y}
\end{equation}
    
   Taking the first derivetive with respect to $X_1$ and $X_2$, we have:
    
\begin{equation}
    X_1=\frac{\alpha}{p_1}\overline{Y}\quad \text{and} \quad X_2=\frac{1-\alpha}{p_2}\overline{Y}
\end{equation}

   In the production sector firms maximize their profits $\Pi_i$ given their techonology constriant as well as producer factor prices:
\begin{equation}
    \max\limits_{L_i,K_i} \Pi_i=q_iY_i-wL_i-rK_i\quad s.t. \quad Y_i=L_i^{\beta_i}K_i^{1-\beta_i}
\end{equation}
    
   Taking first derivatives yields the input demand functions for firms producing good i:
\begin{equation}
L_i=\frac{\beta_i}{w}q_iY_i\quad\text{and}\quad K_i=\frac{1-\beta_i}{r}q_iY_i
\end{equation}

   Note here we assume $q_i=p_i$. In the market economy, goods and factor prices ajust until supply equalds demand in all markets:
\begin{equation}
X_i=Y_i,\quad i=1,2, \quad L_1+L_2=\overline{L}, \quad \text{and} \quad K_1+K_2=\overline{K}
\end{equation}

   Because only the relative prices affects the rquilibrium, we assume $q_1=p_1=1$. Bysubstituting factor demand functions and consumer demand functions and rearranging the goods and factor maket equilibrium consitions, we have:
\begin{align*}
    \frac{1}{p_1}-\left(\frac{\beta_1}{w}\right)^{\beta_1}\left(\frac{1-\beta_1}{r}\right)^{1-\beta_1} & =0 \\
    \frac{1}{p_2}-\left(\frac{\beta_2}{w}\right)^{\beta_2}\left(\frac{1-\beta_2}{r}\right)^{1-\beta_2} & =0 \\
    \frac{\beta_1}{w}\alpha\overline{Y}+\frac{\beta_2}{w}(1-\alpha)\overline{Y}-\overline{L} & = 0
\end{align*}

   Note these three equations imlicitly implies the clear of the capital market. This specifict feature of our system is called *Walras' law*. It says that if $n-1$ markets out of $n$ markets are in equilibrium, then the last market is automatically balanced. The next program uses 'scipy.optimize.minimize' to calculate the optimal solution.
   Note that we cannot use 'scipy.optimize.root' and 'scipy.optimize.fsolve' to find root. The reason is that we need to gurantee our solution is positive. Otherwise we would encounter dividing by zero problem when we use 'scipy.optimize.root'. Equivalently, we can rewrite the obejevtive function system as the absolut value of the function system. The root finding problem then becomes minimizing the absolute value of the objective function system. 'scipy.optimize.minimize' enable us to bound our solution.

```python
##### Program 3.2 Market solution to the static equilibrium model

# Import package
import numpy as np
from scipy.optimize import minimize

# Parameters
Kbar = 10 # Total capital
Lbar = 20 # Total labour
α = 0.3
β = np.array([0.3,0.6]) 
p = np.array([1.0,0.1]) 
# The first element of the price array is p1=1 (constant), the second is p2 which needs to be optimized.

# Defining the objective function system
def markets(x):
    # prices (excluded p1=1), x is a 3-dimentional array
    p[1] = x[0] # p2 is the first element of input x
    w    = x[1] # wage w is the second element of input x
    r    = x[2] # interest rate r is the third elemendt of input x
    # total income
    Ybar = w * Lbar + r * Kbar
    # get market equation
    sol = np.zeros(3)
    sol[0] = 1/p[0]-(β[0]/w)**β[0]*((1-β[0])/r)**(1-β[0]) # Market of good 1 clear
    sol[1] = 1/p[1]-(β[1]/w)**β[1]*((1-β[1])/r)**(1-β[1]) # Market of good 2 clear
    sol[2] = β[0]*α*Ybar/w + β[1]*(1-α)*Ybar/w - Lbar # Labour market clear
    return sol

# Variable bounds
eps = 1.0e-8 # eps is a number close to zero (but not equal to zero) as we need to gurantee our solution >0
bnds = [(eps, None) for i in range(3) ] # defining our bound intervals as (0, inf), (0, inf), (0, inf)

# initial point
x0 = np.zeros(3) + 5

# Defining the function of absolute value of the objective funtion
def absolute(x):
    return np.linalg.norm(markets(x))

# Solve the minimization problem
res = minimize(absolute, x0, bounds=bnds)
print(res)

p[1] = res.x[0] # p2 is the first element of the solution
w    = res.x[1] # w is the second element of the solution
r    = res.x[2] # r is the third element of the solution

# Calculate other economic variables
Ybar = w * Lbar + r * Kbar
Y[0] = α * Ybar / p[0]
Y[1] = (1 - α) * Ybar / p[1]
L    = β * p * Y / w
K    = (np.ones(2) - β) * p * Y / r
U    = Y[0] ** α * Y[1] ** (1 - α)

# Output 
print('Goods Market 1 :')
print('X1 = ', Y[0], 'Y1 = ', Y[0], 'q1 = ', p[0], 'p1 = ', p[0])
print('Goods Market 2 :')
print('X2 = ', Y[1], 'Y2 = ', Y[1], 'q2 = ', p[1], 'p2 = ', p[1])
print('Labour Market :')
print('L1 = ', L[0], 'L2 = ', L[1], 'L = ', Lbar, 'w = ', w)
print('Capital Market :')
print('K1 = ', K[0], 'K2 = ', K[1], 'K = ', Kbar, 'r = ', r)
print('Utility : ')
print('U = ', U)
```

# 3.1.3 Variable labour supply
In this subsection we introduce a labour-supply decision on the household side.

   The household maximizes utility by choosing consumption demand $X_i$ and leisure $l$, keeping in mind the budget and time constraint.
\begin{equation}
\max\limits_{X_1,X_2,l} X_1^{\alpha_1}X_2^{\alpha_2}l^{1-\alpha_1-\alpha_2}\quad \text{s.t.} \quad p_1X_1+p_2X_2+wl=w\overline{T}+r\overline{K}:=\overline{Y}
\end{equation}
    
   where $\overline{T}$ denotes the time endowment fo the household. As before, we have the following function system:
\begin{align*}
    \frac{1}{p_1}-\left(\frac{\beta_1}{w}\right)^{\beta_1}\left(\frac{1-\beta_1}{r}\right)^{1-\beta_1} & =0 \\
    \frac{1}{p_2}-\left(\frac{\beta_2}{w}\right)^{\beta_2}\left(\frac{1-\beta_2}{r}\right)^{1-\beta_2} & =0 \\
    \frac{\beta_1}{w}\alpha\overline{Y}+\frac{\beta_2}{w}(1-\alpha)\overline{Y}+\frac{1-\alpha_1-\alpha_2}{w}-\overline{T} & = 0
\end{align*}

   Note here that $L_1+L_2=\overline{T}-l$. We use the fowlling program to find the root of this system.
    

```python
##### Program 3.3 Model with variable labour supply

# Import package
import numpy as np
from scipy.optimize import minimize

# Parameters
Kbar = 10 # Total capital
Tbar = 30 # Total time endowment
α = np.array([0.3,0.4]) # Set the α1 and α2
β = np.array([0.3,0.6]) 
p = np.array([1.0,0.1]) 
Y = np.zeros(2)
# The first element of the price array is p1=1 (constant), the second is p2 which needs to be optimized.

# Defining the objective function system
def markets(x):
    # prices (excluded p1=1), x is a 3-dimentional array
    p[1] = x[0] # p2 is the first element of input x
    w    = x[1] # wage w is the second element of input x
    r    = x[2] # interest rate r is the third elemendt of input x
    # total income
    Ybar = w * Tbar + r * Kbar
    # get market equation
    sol = np.zeros(3)
    sol[0] = 1/p[0]-(β[0]/w)**β[0]*((1-β[0])/r)**(1-β[0]) # Market of good 1 clear
    sol[1] = 1/p[1]-(β[1]/w)**β[1]*((1-β[1])/r)**(1-β[1]) # Market of good 2 clear
    sol[2] = β[0]*α[0]*Ybar/w + β[1]*α[1]*Ybar/w + (1-α[0]-α[1])*Ybar/w - Tbar # Labour market clear
    return sol

# Variable bounds
eps = 1.0e-8 # eps is a number close to zero (but not equal to zero) as we need to gurantee our solution >0
bnds = [(eps, None) for i in range(3) ] # defining our bound intervals as (0, inf), (0, inf), (0, inf)

# Initial point
x0 = np.zeros(3) + .5

# Defining the function of absolute value of the objective funtion
def absolute(x):
    return np.linalg.norm(markets(x))

# Solve the minimization problem
res = minimize(absolute, x0, bounds=bnds)

p[1] = res.x[0] # p2 is the first element of the solution
w    = res.x[1] # w is the second element of the solution
r    = res.x[2] # r is the third element of the solution

# Calculate other economic variables
Ybar = w * Tbar + r * Kbar
Y[0] = α[0] * Ybar / p[0]
Y[1] = α[1] * Ybar / p[1]
ell  = (1 - α[0] - α[1]) * Ybar / w # calculate the leisure choice
L    = β * p * Y / w
K    = (np.ones(2) - β) * p * Y / r
U    = Y[0] ** α[0] * Y[1] ** α[1] * ell ** (1 - α[0] - α[1])

# Output 
print('Leisure :')
print('l = ',ell)
print('Goods Market 1 :')
print('X1 = ', Y[0], 'Y1 = ', Y[0], 'q1 = ', p[0], 'p1 = ', p[0])
print('Goods Market 2 :')
print('X2 = ', Y[1], 'Y2 = ', Y[1], 'q2 = ', p[1], 'p2 = ', p[1])
print('Labour Market :')
print('L1 = ', L[0], 'L2 = ', L[1], 'T - F = ', Tbar - ell, 'w = ', w)
print('Capital Market :')
print('K1 = ', K[0], 'K2 = ', K[1], 'K = ', Kbar, 'r = ', r)
print('Utility : ')
print('U = ', U)
```

# 3.1.4 Public sector and tax incidence analysis
In this subsection, we introduce governemnt activity into our mdoel. The governemnt provides an exogenous level of the public good $G$ and levies consumption and income taxes in order to finace it. For simplicity it is assumed that the public good is equivalent to with Good 1. Consequently, one has to distinguish between consumer and producer prices as well as between gross and net factor prices.
\begin{align*}
p_i=q_i(1+\tau_i),\quad w^n=w(1-\tau_w),\quad \text{and} \quad r^n=r(1-\tau_r)
\end{align*}
where $\tau_i$ denote consumption tax rates and $\tau_w,\tau_r$ define wage and interest income tax rates, respectively. Conumsers' budget constraints is noe defined by:
\begin{equation}
P_1X_1+p_2X_2+w^nl=w^n\overline{T}+r^n\overline{K}:=\overline{Y}^n
\end{equation}

The new consumption and leisure demand functions are:
\begin{equation}
    X_1=\frac{\alpha}{p_1}\overline{Y}^n\quad \text{and} \quad X_2=\frac{\alpha_2}{p_2}\overline{Y}^n\quad \text{and}\quad l=\frac{1-\alpha_1-\alpha_2}{w^n}\overline{Y}^n
\end{equation}

The market equilibrium condition in the market for consumption Good 1 changes to $X_1+G=Y_1$. The equilibrium condition on the second goods market remains unchanged. Consequently, the two goods market equilibrium conditions change to:
\begin{align*}
\frac{\alpha_1\overline{Y}^n}{p_1}+G-\left(\frac{\beta_1}{w}\right)^{\beta_1}\left(\frac{1-\beta_1}{r}\right)^{1-\beta_1}q_1\left(\frac{\alpha_1\overline{Y}^n}{p_1}+G\right) & = 0 \\
1-\left(\frac{\beta_2}{w}\right)^{\beta_2}\left(\frac{1-\beta_2}{r}\right)^{1-\beta_2}q_2 & =0 \\
\end{align*}

The labour market condition changes to:
\begin{equation}
\frac{\beta_1}{w}q_1\left(\frac{\alpha_1\overline{Y}^n}{p_1}+G\right)+\frac{\beta_2}{w}q_2\frac{\alpha_2\overline{Y}^n}{p_2} +\frac{1-\alpha_1-\alpha_2}{w^n}\overline{Y}^n-\overline{T}  = 0
\end{equation}

Finally, the government budget equilibrium condition is:
\begin{equation}
q_1G-\sum^2_{i=1}\frac{\tau_i}{1+\tau_i}\alpha_i\overline{Y}^n-\tau_ww\left(\overline{T}-\frac{1-\alpha_1-\alpha_2}{w^n}\overline{Y}^n\right)-\tau_rr\overline{K}=0
\end{equation}

The next program finds the root of the function system containing above four equations. Note here only one tax rates are engodenous while other three are exogenous. The next program show the example that only $\tau_r$ is endogenous and other three tax rates equal to zero.

```python
##### Program 3.4 Model with government activity

# Import package
import numpy as np
from scipy.optimize import minimize

# Parameters
Kbar = 10
Tbar = 30
α = np.array([0.3,0.4])
β = np.array([0.3,0.6])
G = 3
τw = 0
τc = np.zeros(2)

# Engodenous variables
τr = 0
K = np.zeros(2)
L = np.zeros(2)
p = np.ones(2) # consumer price
q = np.ones(2) # producer price
sol = np.zeros(4)
Y = np.zeros(2)

# Function to determine market equilibrium
def markets(x):
    # copy prices
    q[1] = x[0]
    w    = x[1]
    r    = x[2]
    τr   = x[3] # set tax rates
#     τw   = x[3]
#     τc[0]= x[3]
#     τc[1]= x[3]
    # calculate consumer prices and total income
    p    = q * (1 + τc)
    wn   = w * (1 - τw)
    rn   = r * (1 - τr)
    Ybarn = wn * Tbar + rn * Kbar
    # get market equation
    sol[0] = α[0]*Ybarn/p[0]+G-(β[0]/w)**β[0]*((1-β[0])/r)**(1-β[0])*q[0]*(α[0]*Ybarn/p[0]+G)
    sol[1] = 1-(β[1]/w)**β[1]*((1-β[1])/r)**(1-β[1])*q[1]
    sol[2] = β[0]/w*q[0]*(α[0]*Ybarn/p[0]+G)+β[1]/w*q[1]*α[1]*Ybarn/p[1]+(1-α[0]-α[1])*Ybarn/wn-Tbar
    sol[3] = q[0]*G-τc[0]/(1+τc[0])*α[0]*Ybarn-τc[1]/(1+τc[1])*α[1]*Ybarn-τw*w*(Tbar-(1-α[0]-α[1])/wn*Ybarn)-τr*r*Kbar
    return sol

# Variable bounds
eps = 1.0e-8 # eps is a number close to zero (but not equal to zero) as we need to gurantee our solution >0
bnds = [(eps, None) for i in range(4) ] # defining our bound intervals as (0, inf), (0, inf), (0, inf), (0,inf)
# Note, tax rates can be negative if we have multiple tax rates

# Find market equilibrium
# Initial point
x0 = np.zeros(4) + .5

# Defining the function of absolute value of the objective funtion
def absolute(x):
    return np.linalg.norm(markets(x))

# Solve the minimization problem
res = minimize(absolute, x0, bounds=bnds)

print(res)
p[1] = res.x[0]
w    = res.x[1]
r    = res.x[2]
τr   = res.x[3]

# calculate consumer prices and total income
p    = q * (1 + τc)
wn   = w * (1 - τw)
rn   = r * (1 - τr)

# Calculate other economic variables
Ybarn = wn * Tbar + rn * Kbar
Xd   = α*Ybarn / p
Y[0] = Xd[0] + G
Y[1] = Xd[1]
ell  = (1 - α[0] - α[1]) * Ybarn / wn # calculate the leisure choice
L    = β * q * Y / w
K    = (np.ones(2) - β) * q * Y / r
U    = Xd[0] ** α[0] * Xd[1] ** α[1] * ell ** (1 - α[0] - α[1])

# Output 
print('Leisure :')
print('l = ',ell)
print('Goods Market 1 :')
print('X1 = ', Xd[0], 'G  = ', G, 'Y1 = ', Y[0])
print('q1 = ', q[0], 'p1 = ', p[0], 'tc1 = ', τc[0])
print('Goods Market 2 :')
print('X2 = ', Xd[1], 'G  = ', 0, 'Y2 = ', Y[1])
print('q2 = ', q[1], 'p2 = ', p[1], 'tc2 = ', τc[1])
print('Labour Market :')
print('L1 = ', L[0], 'L2 = ', L[1], 'T-F = ', Tbar-ell)
print('w = ', w, 'wn = ', wn, 'tw = ', τw)
print('Capital Market :')
print('K1 = ', K[0], 'K2 = ', K[1], 'K = ', Kbar)
print('r = ', r, 'rn = ', rn, 'tr = ', τr)
print('Government (revenue):')
print('tc1 = ', τc[0]*q[0]*Xd[0], 'tc2 = ', τc[1]*q[1]*Xd[1])
print('tw = ', τw*w*(Tbar-ell), 'tr = ', τr*r*Kbar, 'G = ', q[0]*G)
print('Utility : ')
print('U = ', U)
```

# 3.2 Extensions of the basic model
# 3.2.1 Impefect labour markets and unemployment policy
In this subsection, we introduce unemployment in our present model. A very simple specification of the wage cure is a log-linear function such as:
\begin{equation}
\log\left(\frac{w^n}{P}\right)=\gamma_1-\gamma_2\log(ur)
\end{equation}
where $w^n=w(1-\tau_w-\tau_u)$. $\tau_u=\frac{P\overline{B}ur}{w(1-ur)}$ is the unemployment contribution and $P=\frac{\alpha_1}{\alpha_1+\alpha_2}p_1+\frac{\alpha_2}{\alpha_1+\alpha_2}p_2$ denotes the consumer price index.

From the individual perspective, unemployment risk makes income uncertain. Therefore, the unemployment rate is to compute expected income. The budget constraint changes to:
\begin{align*}
p_1X_1+p_2X_2&=w^nL^S(1-ur)+P\overline{B}L^Sur+r^n\overline{K} \\
& = (\overline{T}-l)[w^n(1-ur)+P\overline{B}ur]+r^n\overline{K}
\end{align*}
Let $\tilde{w}^n=w^n(1-ur)+P\overline{B}ur$ denotes the expected net wage and $\overline{B}$ denotes the nominal unemployment benefit, consumers' budget constraint is given by:
\begin{equation}
p_1X_1+p_2X_2+\tilde{w}^nl=\tilde{w}^n\overline{T}+r^n\overline{K}:=\tilde{Y}
\end{equation}

Consequently, the function system changes to:
\begin{align*}
\frac{\alpha_1\tilde{Y}^n}{p_1}+G-\left(\frac{\beta_1}{w}\right)^{\beta_1}\left(\frac{1-\beta_1}{r}\right)^{1-\beta_1}q_1\left(\frac{\alpha_1\tilde{Y}^n}{p_1}+G\right) & = 0 \\
1-\left(\frac{\beta_2}{w}\right)^{\beta_2}\left(\frac{1-\beta_2}{r}\right)^{1-\beta_2}q_2 & =0 \\
\frac{\beta_1}{w}\alpha_1\tilde{Y}^n+\frac{\beta_2}{w}\alpha_2\tilde{Y}^n + (1-ur)\left(\frac{1-\alpha_1-\alpha_2}{\tilde{w}^n}\tilde{Y}^n-\overline{T}\right) & = 0 \\
q_1G-\sum^2_{i=1}\frac{\tau_i}{1+\tau_i}\alpha_i\tilde{Y}^n-\tau_ww\left(\overline{T}-\frac{1-\alpha_1-\alpha_2}{w^n}\tilde{Y}^n\right)-\tau_rr\overline{K} &= 0 \\
\log\left(\frac{w^n}{P}\right)-\gamma_1+\gamma_2\log(ur)&=0
\end{align*}

Next program finds the root with $\tau_c=\tau_w=0,G=3,\overline{B}=0.1,\gamma_1=-2.5,$ and $\gamma_2=0.5$.

```python
##### Program 3.5 Model with unemployment

# Import package
import numpy as np
from scipy.optimize import fmin, minimize,root

# Parameters
Kbar = 10
Tbar = 30
α = np.array([0.3,0.4])
β = np.array([0.3,0.6])
γ = np.array([-2.5,0.5])
G = 3
B = 0.1
τw = 0
τc = np.zeros(2)

# Endogenous variable
τu = 0 # Set the unemployment contribution
τr = 0 # Set the capital tax
ur = 0 # Set the unemployment rate
px = 0 # Set the consumers price index
K = np.zeros(2)
L = np.zeros(2)
p = np.ones(2)
q = np.ones(2)
sol = np.zeros(5)
Y = np.zeros(2)

# Function to determine market equilibrium
def markets(x):
    # copy prices
    q[1] = x[0]
    w    = x[1]
    r    = x[2]    
    # set tax rates
    τr   = x[3]
    # copy the unemployment rate
    ur   = x[4]
    # calculate consumer prices and total income
    p    = q * (1 + τc)
    px   = p[0] * α[0] / np.sum(α) + p[1] * α[1] / np.sum(α)
    τu   = px * B * ur / (w * (1 - ur))
    wn   = w*(1-τw-τu)
    wt   = wn * (1 - ur) + px * B * ur
    rn   = r * (1 - τr)
    Ytn = wt * Tbar + rn * Kbar
    # get market equation
    sol[0] = α[0]*Ytn/p[0]+G-(β[0]/w)**β[0]*((1-β[0])/r)**(1-β[0])*q[0]*(α[0]*Ytn/p[0]+G)
    sol[1] = 1-(β[1]/w)**β[1]*((1-β[1])/r)**(1-β[1])*q[1]
    sol[2] = β[0]*q[0]*(α[0]*Ytn/p[0]+G)/w+β[1]*q[1]*α[1]*Ytn/(w*p[1])+(1-ur)*((1-α[0]-α[1])*Ytn/wt-Tbar)
    sol[3] = q[0]*G-τc[0]/(1+τc[0])*α[0]*Ytn-τc[1]/(1+τc[1])*α[1]*Ytn-τw*w*(Tbar-(1-α[0]-α[1])*Ytn/wt)-τr*r*Kbar
    sol[4] = (wn/px)*ur**γ[1]-np.exp(γ[0])
    return sol

# Find market equilibrium
# Initial point
x0 = np.zeros(5) + .5

# Variable bounds
eps = 1.0e-8 # eps is a number close to zero (but not equal to zero) as we need to gurantee our solution >0
bnds = [(eps, None) for i in range(5) ] 
# defining our bound intervals as (0, inf), (0, inf), (0, inf), (0,inf), (0,inf) 
# Note, tax rates can be negative if we have multiple tax rates

# Defining the function of absolute value of the objective funtion
def absolute(x):
    return np.linalg.norm(markets(x))

# Solve the minimization problem
res = minimize(absolute, x0, bounds=bnds)

print(res)
q[1] = res.x[0]
w    = res.x[1]
r    = res.x[2]
τr   = res.x[3]
ur   = res.x[4]

# calculate consumer prices and total income
p    = q * (1 + τc)
px   = p[0] * α[0] / sum(α) + p[1] * α[1] / sum(α)
τu   = px * B * ur / (w * (1 - ur))
wn   = w * (1 - τw - τu)
wt   = wn * (1 - ur) + px * B * ur
rn   = r * (1 - τr)

# Calculate other economic variables
Ytn = wt * Tbar + rn * Kbar
Xd   = Ytn*α/p
Y[0] = Xd[0] + G
Y[1] = Xd[1]
ell  = (1 - α[0] - α[1]) * Ytn / wt
L    = β * q * Y / w
K    = (np.ones(2) - β) * q * Y / r
U    = Xd[0] ** α[0] * Xd[1] ** α[1] * ell ** (1 - α[0] - α[1])

# Output 
print('Leisure :')
print('l = ',ell)
print('Unemployment:')
print('ur = ',ur, 'τu = ',τu)
print('Goods Market 1 :')
print('X1 = ', Xd[0], 'G  = ', G, 'Y1 = ', Y[0])
print('q1 = ', q[0], 'p1 = ', p[0], 'tc1 = ', τc[0])
print('Goods Market 2 :')
print('X2 = ', Xd[1], 'G  = ', 0, 'Y2 = ', Y[1])
print('q2 = ', q[1], 'p2 = ', p[1], 'tc2 = ', τc[1])
print('Labour Market :')
print('L1 = ', L[0], 'L2 = ', L[1], 'T-F = ', Tbar-ell) ## ??? L1+L2 does not equal T-F due to rounding error ???
print('w = ', w, 'wt = ', wt, 'tw = ', τw)
print('Capital Market :')
print('K1 = ', K[0], 'K2 = ', K[1], 'K = ', Kbar)
print('r = ', r, 'rn = ', rn, 'tr = ', τr)
print('Government (revenue):')
print('tc1 = ', τc[0]*q[0]*Xd[0], 'tc2 = ', τc[1]*q[1]*Xd[1])
print('tw = ', τw*w*(Tbar-ell), 'tr = ', τr*r*Kbar, 'G = ', q[0]*G)
print('Utility : ')
print('U = ', U)
```

```python
##### Program 3.6 Model with intermediate goods

# Import package
import numpy as np
from scipy.optimize import fmin, minimize,root

# Parameters
Kbar = 10
Tbar = 30
α = np.array([0.3,0.4])
β = np.array([0.3,0.6])
a0 = np.array([0.2,0.2])
a1 = np.array([[1.0,-0.2],[-0.3,1.0]]) # The first linear sysmtem (for computing q1 and q2)
a2 = np.array([[1.0,-0.3],[-0.2,1.0]]) # The second linear system (for conputing Y1 and Y2)
w  = 1 # Normalize wage to w = 1
G = 3
τw = 0 # Assume no income tax
τr = 0 # Assume no capital tax

# Endogenous Variables
Yb = np.zeros(2)
qb = np.zeros(2)

# Function to determine market equilibrium
def markets(x):
    global q
    # copy prices
    r    = x[0]
    τc[0]= x[1]
    τc[1]= τc[0]    
    
    # calculate K/Y and L/Y
    ky   = a0*((((np.ones(2)-β)/β)*(w/r))**β)
    ly   = a0*((β/(np.ones(2)-β))*r/w)**(np.ones(2)-β)
    
    # determin producer prices
    qb   = w * ly + r * ky # Calculate the RHS of the linear system of q
    q    = np.linalg.solve(a1,qb)
    
    # calculate consumer prices and total income
    p    = q * (1 + τc)
    wn   = w * (1 - τw)
    rn   = r * (1 - τr)
    Ybarn = wn * Tbar + rn * Kbar
    Xd   = α / p * Ybarn
    ell  = (1 - α[0] - α[1])/ wn * Ybarn
    
    # determine output levels
    Yb[0] = Xd[0] + G
    Yb[1] = Xd[1]
    Y = np.linalg.solve(a1,Yb)
    
    # compute K and L
    K = ky*Y
    L = ly*Y
    # check markets and budget
    sol[0] = K[0] + K[1] - Kbar
    sol[1] = q[0]*G-np.sum(τc*q*Xd)-τw*w*(Tbar-ell)-τr*r*Kbar
    return sol

# Find market equilibrium
# Initail Guess
x0 = ([0.2,0.5])

# Variable bounds
eps = 1.0e-8 # eps is a number close to zero (but not equal to zero) as we need to gurantee our solution >0
bnds = [(eps, None),(eps, None)] 
# defining our bound intervals as (0, inf), (0, inf)
# Note, tax rates can be negative if we have multiple tax rates

# Defining the function of absolute value of the objective funtion
def absolute(x):
    return np.linalg.norm(markets(x))

# Solve the minimization problem
res = minimize(absolute, x0, bounds=bnds)
print(res)

r = res.x[0]
τc[0] = res.x[1]

# calculate consumer prices and total income
p    = q * (1 + τc)
wn   = w * (1 - τw)
rn   = r * (1 - τr)

# Calculate other economic variables
Ybarn = wn * Tbar + rn * Kbar
Xd   = Ybarn*α / p
Y[0] = Xd[0] + G
Y[1] = Xd[1]
ell  = (1 - α[0] - α[1]) * Ybarn / wn
L    = β * q * Y / w
K    = (np.ones(2) - β) * q * Y / r
U    = Xd[0] ** α[0] * Xd[1] ** α[1] * ell ** (1 - α[0] - α[1])

# Output 
print('Goods Market 1 :')
print('X11 = ', (1-a1[0,0])*Y[0], 'X12 = ', -a1[1,0]*Y[1], 'q1X1 = ', q[0]*Xd[0], 'q1G = ', 
      q[0]*G, 'q1Y1 = ',q[0]* Y[0])
print('q1 = ', q[0], 'p1 = ', p[0], 'tc1 = ', τc[0])
print('---------------------------')
print('Goods Market 2 :')
print('X21 = ', -a1[0,1]*Y[0], 'X22 = ', (1-a1[1,1])*Y[1], 'q2X2 = ', q[1]*Xd[1], 'Y2 = ', q[1]* Y[1])
print('q2 = ', q[1], 'p2 = ', p[1], 'tc2 = ', τc[1])
print('---------------------------')
print('Labour Market :')
print('L1 = ', L[0], 'L2 = ', L[1], 'T-F = ', Tbar-ell)
print('w = ', w, 'wn = ', wn, 'tw = ', τw)
print('---------------------------')
print('Capital Market :')
print('K1 = ', K[0], 'K2 = ', K[1], 'K = ', Kbar)
print('r = ', r, 'rn = ', rn, 'tr = ', τr)
print('---------------------------')
print('Utility : ')
print('U = ', U)
```
